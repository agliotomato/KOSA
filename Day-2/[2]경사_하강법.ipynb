## **경사 하강법을 이용한 최소 값 확인**

### 경사 하강법 수식

```python
def gradient_descent(my_function, x, delta, lr) :
  # 코드를 작성해주세요
  gradient = derivative(my_function,x, delta)
  new_x = x - lr*gradient
  return new_x
```

#### 정답

```python
def gradient_descent(my_function, x, delta, lr) :
  gradient = derivative(my_function, x, delta)
  new_x = x - lr*gradient
  return new_x
```

### 시작점

```python
init_x = -1.5
```

### **경사 하강법을 적용하여 최소 값 확인**

```python
new_x_list, new_y_list = [], []

start_x = init_x
iteration = 10
lr = 1e-1

for i in range(iteration) :
  new_x = gradient_descent(my_function, start_x, delta, lr)
  new_x_list.append(new_x)
  new_y_list.append(my_function(new_x))
  start_x = new_x
```

### **경사 하강 과정 시각화**

```python
plt.figure(figsize=(10, 5))

region_x, region_y = np.arange(init_x, 0, delta), my_function(np.arange(init_x, 0, delta))
plt.plot(region_x, region_y)

plt.scatter(new_x_list, new_y_list)

for i in range(iteration - 1) :
  delta_x, delta_y = new_x_list[i+1] - new_x_list[i], new_y_list[i+1] - new_y_list[i]
  plt.arrow(new_x_list[i], new_y_list[i], delta_x, delta_y, head_width=0.05, head_length=0.1, fc="k", ec="k")

plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Gradient Descent")
plt.show()
```

## **경사 하강법에 기반한 선형 회귀 분석**

### 정량적 평가를 위한 Scikit-Learn 라이브러리 사용

```python
from sklearn.metrics import r2_score
```

### 선형 모델 함수 샘플링

```python
def my_function(x) :
  return 10*x + 5
```

### 선형 모델 함수 시각화

```python
x = np.arange(0, 10, 1e-2)
y = my_function(x)
```

```python
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("my_function")
plt.show()
```

### **경사 하강법 기반 매개변수 최신화 함수 (MSE 손실 함수)**

```python
def gradient_descent(input, target, a, b, lr) :
  # 코드를 작성해주세요
  gradient_a = 2*(a*np.power(input, 2) + b*input - input*target).mean()
  gradient_b = 2*(a*input + b - target).mean()

  new_a = a - lr*gradient_a
  new_b = b - lr*gradient_b
  return new_a, new_b
```

#### 정답

```python
def gradient_descent(input, target, a, b, lr) :
  gradient_a = 2*(a*np.power(input, 2) + b*input - input*target).mean()
  gradient_b = 2*(a*input + b - target).mean()

  new_a = a - lr*gradient_a
  new_b = b - lr*gradient_b

  return new_a, new_b
```

### 시드 고정

```python
np.random.seed(42)
```

### 모델 파라미터 (매개변수) 초기화

```python
a, b = np.random.random(), np.random.random()
print(a, b)
```

### 하이퍼파라이터 선정

```python
iteration, lr = 100, 1e-2
```

### **경사 하강을 통한 파라미터 최신화**

```python
old_a, old_b = a, b

list_a, list_b = [], []
list_loss = []

for i in range(iteration) :
  pred = old_a * x + b
  target = my_function(x)

  loss = np.power((pred - target), 2).mean()
  list_loss.append(loss)

  new_a, new_b = gradient_descent(x, y, old_a, old_b, lr)
  old_a, old_b = new_a, new_b

  list_a.append(old_a)
  list_b.append(old_b)
```

### 손실 함수 시각화

```python
plt.figure(figsize=(10, 5))
plt.plot(np.arange(iteration), list_loss)
plt.xlabel("Iteration")
plt.ylabel("MSE Loss")
plt.title("Loss Graph")
plt.show()
```

### **최신화 된 모델 파라미터 출력**

```python
final_a, final_b = list_a[-1], list_b[-1]
print(final_a, final_b)
```

### **모델 파라미터를 기준으로 결과 추론**

```python
y_pred = final_a*x + final_b
```

### **추론 결과의 정성적 및 정략성 평가**

```python
plt.figure(figsize=(10, 5))
plt.plot(x, y, label="Ground Truth")
plt.plot(x, y_pred, label="Prediction")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Regression Result")
plt.legend(loc="best")
plt.show()
```

```python
r2_score(y, y_pred)
```

### **훈련된 모델 파라미터의 외삽법 (Extrapolation) 결과**

```python
x = np.arange(20, 30, 1e-2)
```

```python
y = my_function(x)
```

```python
y_pred = final_a*x + final_b
```

```python
plt.figure(figsize=(10, 5))
plt.plot(x, y, label="Ground Truth")
plt.plot(x, y_pred, label="Prediction")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Regression Result")
plt.legend(loc="best")
plt.show()
```

```python
r2_score(y, y_pred)
```

## **경사 하강법에 기반한 다중 레이어 모델 기반 회귀 분석**

### 비선형 모델 함수 샘플링

```python
def my_function(x) :
  return 4*np.power(x, 2) + 2*x + 1
```

### 비선형 모델 함수 시각화

```python
x = np.arange(0, 10, 1e-2)
y = my_function(x)
```

```python
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("my_function")
plt.show()
```

### **경사 하강법 기반 매개변수 최신화 함수 (MSE 손실 함수)**

```python
def f(x, a, b) :
  return a*x + b

def g(my_function, x, c, d) :
  return c*x*my_function + d
```

```python
def gradient_descent(input, target, a, b, c, d, lr) :
  # 코드를 작성해주세요
  gradient_a = 2*((g(f(input, a, b), input, c, d) - target) * (c*input + d) * input).mean()
  gradient_b = 2*((g(f(input, a, b), input, c, d) - target) * (c*input + d)).mean()
  gradient_c = 2*((g(f(input, a, b), input, c, d) - target) * input * f(input, a, b)).mean()
  gradient_d = 2*((g(f(input, a, b), input, c, d) - target)).mean()

  new_a = a - lr*gradient_a
  new_b = b - lr*gradient_b
  new_c = c - lr*gradient_c
  new_d = d - lr*gradient_c
  return new_a, new_b, new_c, new_d
```

#### 정답

```python
def gradient_descent(input, target, a, b, c, d, lr) :
  gradient_a = 2*((g(f(input, a, b), input, c, d) - target) * (c*input + d) * input).mean()
  gradient_b = 2*((g(f(input, a, b), input, c, d) - target) * (c*input + d)).mean()
  gradient_c = 2*((g(f(input, a, b), input, c, d) - target) * input * f(input, a, b)).mean()
  gradient_d = 2*((g(f(input, a, b), input, c, d) - target)).mean()

  new_a = a - lr*gradient_a
  new_b = b - lr*gradient_b
  new_c = c - lr*gradient_c
  new_d = d - lr*gradient_c

  return new_a, new_b, new_c, new_d
```

### 모델 파라미터 (매개변수) 초기화

```python
a, b, c, d, = np.random.random(), np.random.random(), np.random.random(), np.random.random()
```

### 하이퍼파라이터 선정

```python
iteration, lr = 2000, 1e-7
```

### **경사 하강을 통한 파라미터 최신화**

```python
old_a, old_b, old_c, old_d = a, b, c, d

list_a, list_b, list_c, list_d = [], [], [], []
list_loss = []

for i in range(iteration) :
  pred = old_c*x*(old_a*x + old_b) + old_d
  target = my_function(x)

  loss = np.power((pred - target), 2).mean()
  list_loss.append(loss)

  new_a, new_b, new_c, new_d = gradient_descent(x, y, old_a, old_b, old_c, old_d, lr)
  old_a, old_b, old_c, old_d = new_a, new_b, new_c, new_d

  list_a.append(old_a)
  list_b.append(old_b)
  list_c.append(old_c)
  list_d.append(old_d)
```

### 손실 함수 시각화

```python
plt.figure(figsize=(10, 5))
plt.plot(np.arange(iteration), list_loss)
plt.xlabel("Iteration")
plt.ylabel("MSE Loss")
plt.title("Loss Graph")
plt.show()
```

### **최신화 된 모델 파라미터 출력**

```python
final_a, final_b, final_c, final_d = list_a[-1], list_b[-1], list_c[-1], list_d[-1]
print(final_a, final_b, final_c, final_d)
```

### **모델 파라미터를 기준으로 결과 추론**

```python
y_pred = final_c*x*(final_a*x + final_b) + final_d
```

### **추론 결과의 정성적 평가**

```python
plt.figure(figsize=(10, 5))
plt.plot(x, y, label="Ground Truth")
plt.plot(x, y_pred, label="Prediction")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Regression Result")
plt.legend(loc="best")
plt.show()
```

### **훈련된 모델 파라미터의 외삽법 (Extrapolation) 결과**

```python
x = np.arange(20, 30, 1e-2)
```

```python
y = my_function(x)
```

```python
y_pred = final_c*x*(final_a*x + final_b) + final_d
```

```python
plt.figure(figsize=(10, 5))
plt.plot(x, y, label="Ground Truth")
plt.plot(x, y_pred, label="Prediction")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Regression Result")
plt.legend(loc="best")
plt.show()
```
