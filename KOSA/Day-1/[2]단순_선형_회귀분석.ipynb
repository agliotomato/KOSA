# **단순 선형 회귀분석**

```python
# 사용할 라이브러리 불러오기
import numpy as np
import matplotlib.pyplot as plt
```

## **선형 방정식을 사용하여 샘플링 진행하기**


### 약한 Noise를 추가하여 상관계수를 조정

```python
np.random.seed(0) # 시드 고정
```

```python
def function(x:np.array, a:int, b:int) :
  output = a*x + b + 0.25*np.random.randn(x.shape[0])

  return output
```

```python
x = np.arange(0, 10, 0.1)
y = function(x, 5, 10)
```

```python
plt.figure(figsize=(10,5))
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Linear Equation")
plt.show()
```

### **단선 선형 회귀 모델의 파라미터 계산**

### **기울기 (Gradient) 계산**

```python
def compute_weight(x:np.array, y:np.array) :
  # 코드를 작성해주세요
  output = np.sum((x - np.mean(x)) * (y - np.mean(y))) / np.sum((x - np.mean(x)) ** 2)
  return output
```

```python
gradient = compute_weight(x, y)
print(gradient)
```

#### 정답

```python
def compute_weight(x:np.array, y:np.array) :
  output = np.sum((x - np.mean(x))*(y - np.mean(y))) / np.sum((x - np.mean(x))**2)
  return output
```

### **절편 (Bias) 계산**



```python
def compute_bias(x:np.array, y:np.array, gradient:float) :
  # 코드를 작성해주세요
  output = np.mean(y) - gradient*np.mean(x)
  return output
```

```python
bias = compute_bias(x, y, gradient)
print(bias)
```

#### 정답

```python
def compute_bias(x:np.array, y:np.array, gradient:float) :
  output = np.mean(y) - gradient*np.mean(x)
  return output
```

### **추론 (Inference) 결과 비교**

```python
y_hat = gradient*x + bias
```

#### 정답

```python
y_hat = gradient*x + bias
```

```python
plt.figure(figsize=(10,5))
plt.plot(x, y, label="ground truth")
plt.plot(x, y_hat, label="prediction")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Linear Equation")
plt.legend()
plt.show()
```
