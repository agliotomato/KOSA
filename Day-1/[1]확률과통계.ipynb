## **샘플들의 통계량**

### **평균**

#### Python 코딩을 통한 평균 계산 구현

```python
def compute_mean(sample_set:np.array) :
  # 코드를 작성해주세요
  mean = np.sum(sample_set)/len(sample_set)
  return mean
```

##### 정답

```python
def compute_mean(sample_set:np.array) :
  mean = np.sum(sample_set)/len(sample_set)
  return mean
```

```python
print(f"Sample Set 0 : {compute_mean(sample_set_0):.4f}")
print(f"Sample Set 1 : {compute_mean(sample_set_1):.4f}")
print(f"Sample Set 2 : {compute_mean(sample_set_2):.4f}")
```

#### Numpy를 사용한 평균 계산

```python
print(f"Sample Set 0 : {np.mean(sample_set_0):.4f}")
print(f"Sample Set 1 : {np.mean(sample_set_1):.4f}")
print(f"Sample Set 2 : {np.mean(sample_set_2):.4f}")
```

### **분산**

#### Python 코딩을 통한 분산 계산 구현
*   앞에서 선언한 compute_mean 함수를 사용하세요!



```python
def compute_var(sample_set:np.array) :
  # 코드를 작성해주세요
  var = np.sum((sample_set - compute_mean(sample_set))** 2)/len(sample_set)
  return var
```

##### 정답

```python
def compute_var(sample_set:np.array) :
  var = np.sum((sample_set - compute_mean(sample_set))**2)/len(sample_set)
  return var
```

```python
print(f"Sample Set 0 : {compute_var(sample_set_0):.4f}")
print(f"Sample Set 1 : {compute_var(sample_set_1):.4f}")
print(f"Sample Set 2 : {compute_var(sample_set_2):.4f}")
```

#### Numpy를 사용한 분산 계산

```python
print(f"Sample Set 0 : {np.var(sample_set_0):.4f}")
print(f"Sample Set 1 : {np.var(sample_set_1):.4f}")
print(f"Sample Set 2 : {np.var(sample_set_2):.4f}")
```

### **표준편차**

#### Python 코딩을 통한 표준편차 계산 구현
*   앞에서 선언한 compute_var 함수를 사용하세요!



```python
def compute_std(sample_set:np.array) :
  # 코드를 작성해주세요
  std = np.sqrt(compute_var(sample_set))
  return std
```

##### 정답

```python
def compute_std(sample_set:np.array) :
  std = np.sqrt(compute_var(sample_set))
  return std
```

```python
print(f"Sample Set 0 : {compute_std(sample_set_0):.4f}")
print(f"Sample Set 1 : {compute_std(sample_set_1):.4f}")
print(f"Sample Set 2 : {compute_std(sample_set_2):.4f}")
```

#### Numpy를 사용한 표준편차 계산

```python
print(f"Sample Set 0 : {np.std(sample_set_0):.4f}")
print(f"Sample Set 1 : {np.std(sample_set_1):.4f}")
print(f"Sample Set 2 : {np.std(sample_set_2):.4f}")
```

### **Pandas DataFrame을 사용하여 결과 정리**

```python
sample_set = {"sample_set_0":sample_set_0, "sample_set_1":sample_set_1, "sample_set_2":sample_set_2}
stats_dict = {"mean":[], "var":[], "std":[], "median":[], "max":[], "min":[]}

for subsample_set in sample_set.values() :
  stats_dict["mean"].append(np.mean(subsample_set))
  stats_dict["var"].append(np.var(subsample_set))
  stats_dict["std"].append(np.std(subsample_set))
  stats_dict["median"].append(np.median(subsample_set))
  stats_dict["max"].append(np.max(subsample_set))
  stats_dict["min"].append(np.min(subsample_set))
```

```python
stats_dict
```

```python
my_df = pd.DataFrame(data=stats_dict)
my_df.index = ["Sample Set 0", "Sample Set 1", "Sample Set 2"]
```

## **선형 방정식을 사용하여 샘플링 진행하기**


```python
def function(x:np.array, a:int, b:int) :
  output = a*x + b
  return output
```

```python
x = np.arange(0, 10, 0.001)
y = function(x, 10, 10)
```

```python
plt.figure(figsize=(10,5))
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Linear Equation")
plt.show()
```

### Numpy를 사용한 공분산 계산

```python
np.cov(x, y)
```

```python
print(f"Covariance : {np.cov(x, y)[0][1]}")
```

### **상관계수**

#### 공식을 이용한 상관계수 계산
*   np.cov 함수를 사용하세요!




```python
def corr_coef(x:np.array, y:np.array) :
  # 코드를 작성해주세요
  cc = np.cov(x,y)[0][1]/(np.std(x)*np.std(y))
  return cc
```

##### 정답

```python
def corr_coef(x:np.array, y:np.array) :
  cc = np.cov(x, y)[0][1] / (np.std(x)*np.std(y))
  return cc
```

```python
corr_coef(x, y)
```

#### Numpy를 사용한 상관계수 계산

```python
np.corrcoef(x, y)
```

```python
np.corrcoef(x, y)[0][1]
```

## **비선형 방정식을 사용하여 샘플링 진행하기**


```python
def function(x:np.array, a:int, b:int, c:int) :
  output = a*np.power(x, 2) + b*x + c

  return output
```

```python
x = np.arange(0, 10, 0.001)
y = function(x, -1, 10, -10)
```

```python
plt.figure(figsize=(10,5))
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Non-Linear Equation")
plt.show()
```

```python
corr_coef(x, y)
```

```python
np.corrcoef(x, y)[0][1]
```
